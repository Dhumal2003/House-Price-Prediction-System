**Data Preprocessing:**

```python
# Check for missing values
df.isnull().sum()
```

```python
#check Duplicate values
df.duplicated().sum()
```

```python
# Check data types of all columns
print(df.dtypes)
```

```python
# Convert number of bethrooms to int
df['number of bathrooms'] = df['number of bathrooms'].astype(int)
```

```python
# Feature Engineering
df['House_Age'] = 2025 - df['Built Year']  # Age of the house
df['Total_Area'] = df['Area of the house(excluding basement)'] + df['Area of the basement']
```

```python
#Drop ID and Date,number of views (not useful for price prediction)
df.drop(columns=['Date','id','number of views'],inplace=True)
```

```python
df['Renovation Year'].value_counts()[1:].sum()
```

```python
df.drop(columns='Renovation Year',inplace=True)
```

```python
df.columns
```

```python
df.drop(columns=['Area of the house(excluding basement)','Area of the basement','living area'],inplace=True)
```

```python
df.to_csv('preprocessed_indian_house_prices1.csv', index=False)
```

```python
data=df
```

```python
# Check updated dataset
data.head()
```

**Data Visualization:**

```python
# Plot the distribution of house prices
plt.figure(figsize=(10, 6))
sns.histplot(data['Price'], kde=True, bins=30)
plt.title('Distribution of House Prices')
plt.xlabel('Price')
plt.ylabel('Frequency')
plt.show()
```

```python
# Correlation heatmap to see feature importance
plt.figure(figsize=(12,8))
sns.heatmap(data.corr(), annot=True, cmap='coolwarm')
plt.show()
```

```python
data.corr()
```

```python
# Split the dataset into features (X) and target (y)
x = data.drop('Price', axis=1)
```

```python
y = data['Price']
y
```

```python
from sklearn.model_selection import train_test_split
```

```python
x_train, x_test, y_train, y_test = train_test_split(x, y, test_size=0.30, random_state=42)
```

```python
x_train.shape,x_test.shape
```

**Standard Scalling:**

```python
from sklearn.preprocessing import StandardScaler
scaler=StandardScaler()
```

```python
x_train=scaler.fit_transform(x_train)  # Scale training features
x_train
```

```python
x_test=scaler.fit_transform(x_test)
x_test
```

**Build and Train Linear Regression Model:**

```python
from sklearn.linear_model import LinearRegression
```

```python
LR=LinearRegression()
```

```python
#Train the model using the scaled data
LR.fit(x_train,y_train)
```

```python
LR.coef_
```

```python
LR.intercept_
```

**Model Prediction:**

```python
y_hat =LR.predict(x_test)
y_hat
```

**Evaluation model:**

```python
from sklearn.metrics  import mean_absolute_error,mean_squared_error,r2_score
```

```python
mae = mean_absolute_error(y_test, y_hat)
mse = mean_squared_error(y_test, y_hat)
r2 = r2_score(y_test, y_hat)

print(f"MAE: {mae}")
print(f"MSE: {mse}")
print(f"R2 Score: {r2}")
```

**Used Log-Transformed to Improve Linear Regression Accuracy:**

```python
#I applied log transformation on target variable 'Price' to reduce skewness.
#it will help the  ML model learn better, and improve accuracy.
```

```python
# Apply log transformation for price
y_train_log = np.log1p(y_train)
y_test_log  = np.log1p(y_test)
```

```python
lr = LinearRegression()
lr.fit(x_train, y_train_log)
```

```python
y_pred_log = lr.predict(x_test)
```

```python
# Convert back to original Price scale
y_pred = np.expm1(y_pred_log)
```

```python
mae = mean_absolute_error(y_test,y_pred)
mse = mean_squared_error(y_test, y_pred)
r2 = r2_score(y_test, y_pred)

print(f"  MAE: {mae:,.2f}")
print(f"  MSE: {mse:,.2f}")
print(f"  R2 : {r2:.4f}")
```

**Hyperparameter Tuning for Lasso Regression (L1 Regularization) using GridSearchCV:**

```python
'''
I used GridSearchCV with Lasso to find the best alpha value automatically, 
which improves model accuracy and avoids manual trial-and-error.

'''
```

```python
from sklearn.linear_model import Lasso
from sklearn.model_selection import GridSearchCV
```

```python

lasso = Lasso()
params = {'alpha': [0.001, 0.01, 0.1, 1, 10]} 
grid = GridSearchCV(lasso, params, cv=5, scoring='r2')
grid.fit(x_train, y_train_log)

best_lasso = grid.best_estimator_  
y_pred_log = best_lasso.predict(x_test)

# Back-transform predictions
y_pred = np.exp(y_pred_log)

print("----- Lasso Regression -----")
print("Best alpha:", grid.best_params_)
print("Best CV R²:", grid.best_score_)
print("Test R²:", r2_score(y_test, y_pred))
```

**Ridge Regression (L2 Regularization):**

```python
from sklearn.linear_model import Ridge
```

```python
#Ridge Regression with GridSearchCV
ridge = Ridge()
params = {'alpha': [0.001, 0.01, 0.1, 1, 10, 100]} 
grid = GridSearchCV(ridge, params, cv=5, scoring='r2')
grid.fit(x_train, y_train_log)

best_ridge = grid.best_estimator_  
y_pred_log = best_ridge.predict(x_test)

print("----- Ridge Regression -----")
print("Best alpha:", grid.best_params_)
print("Best CV R²:", grid.best_score_)
print("Test R²:", r2_score(y_test, y_pred))
```

**Polynomial Regression:**

```python
from sklearn.preprocessing import PolynomialFeatures
```

```python
# Transform features to polynomial features
poly=PolynomialFeatures(degree=2)
x_train_poly=poly.fit_transform(x_train)
x_test_poly=poly.transform(x_test)
```

```python

poly_reg=LinearRegression()
poly_reg.fit(x_train_poly,y_train)
```

```python
y_pred1= poly_reg.predict(x_test_poly)
```

```python
mse = mean_squared_error(y_test, y_pred1)
r2 = r2_score(y_test, y_pred1)

print(f"Mean Squared Error: {mse}")
print(f"R-squared: {r2}")
```

**Using Random Forest Regressor:**

```python
from sklearn.ensemble import RandomForestRegressor
```

```python
RF= RandomForestRegressor(n_estimators=100, random_state=42)
RF.fit(x_train, y_train)
y_pred2 = RF.predict(x_test)
```

```python
mae = mean_absolute_error(y_test, y_pred2)
r2 = r2_score(y_test, y_pred2)

print("Random Forest Performance:")
print(f"  MAE: {mae:,.2f}")
print(f"  R2 : {r2:.4f}")
```

**Model Performance Comparison Chart Using (R² Score):**

```python
results = {
    "Linear Regression (Without using Log)": r2_score(y_test,y_hat),
    "Linear Regression (Using Log)":r2_score(y_test, y_pred),
    "Lasso Regression (Using Hyperparameter Tuning )": r2_score(y_test,y_pred),
    "Ridge Regression (Using Hyperparameter Tuning)": r2_score(y_test,y_pred),
    "Polynomial Regression": r2_score(y_test, y_pred1),
    "Random Forest Regressor": r2_score(y_test, y_pred2)
}
pd.DataFrame.from_dict(results,orient='index', columns=['R² Score'])
```
